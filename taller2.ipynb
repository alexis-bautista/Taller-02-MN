## Parte Proporcionada

```python
def f(x: float, points: list[float] = [], show: bool = False):
    y = x**3 - 3 * x**2 + x - 1
    points.append((x, y))
    if show:
        print(f"f({x}) = {y}")
    return y
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 4, 400)

y = f(x)

plt.figure(figsize=(8 ,6))
plt.plot(x, y, label="$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
from scipy.optimize import newton

x0 = 100
points = []
x_r = newton(
    f, x0, args=(points, True), maxiter=100
)   # Este no es método de Newton, es secante!
```

## Modificaciones para el Taller

Librerias a usar

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton
```

Definir la función y su derivada para el método de Newton

```python
def f(x: float, points: list = None, show: bool = False):
    y = x**3 - 3 * x**2 + x - 1
    if points is not None:
        points.append((x, y))
    if show:
        print(f"f({x}) = {y}")
    return y

def df(x: float, *args):
    return 3 * x**2 - 6 * x + 1  # Derivada de f(x)
```

Dibujo de la ecuacion

```python
x = np.linspace(-2, 4, 400)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$f(x) = x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $f(x) = x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

Funcion para probar varias estimaciones iniciales

```python
def find_root_and_plot(x0, color):
    points = []
    try:
        # Usar el método de Newton con derivada
        root = newton(f, x0, fprime=df, args=(points, False), maxiter=100, full_output=True, disp=False)
        x_r = root[0]
        print(f"Raíz encontrada desde x0 = {x0}: x = {x_r:.4f}")
    except RuntimeError:
        x_r = None
        print(f"No se encontró convergencia desde x0 = {x0}")
        
    # Extraer las estimaciones
    x_points = [p[0] for p in points]
    y_points = [p[1] for p in points]

    # Graficar la trayectoria de las raíces
    plt.plot(x_points, "o-", label=f"Trayectoria desde x0 = {x0}", color=color)
    return x_r
```

Probar tres estimaciones iniciales en la cual una no converge

```python
plt.figure(figsize=(8, 6))

x_r1 = find_root_and_plot(3.0, "blue")  # Converge
x_r2 = find_root_and_plot(0.5, "green") # Converge
x_r3 = find_root_and_plot(-2.0, "red")  # No converge

plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Trayectoria de las estimaciones para el método de Newton")
plt.xlabel("Iteración")
plt.ylabel("Estimación de x")
plt.legend()
plt.show()
```
